```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\AROHI SINGH\Downloads\city_day.csv")
print(df.head(5))
```

```python
df.isnull().sum()
(df.isnull().sum() / len(df)) * 100
```

```python
df.drop(columns=['Xylene'], inplace=True)
```

```python
df['PM2.5'].fillna(df['PM2.5'].median(), inplace=True)
df['PM10'].fillna(df['PM10'].median(), inplace=True)
df['NO2'].fillna(df['NO2'].median(), inplace=True)
df['CO'].fillna(df['CO'].median(), inplace=True)
```

```python
df['AQI'].fillna(df['AQI'].median(), inplace=True)
df['AQI_Bucket'].fillna(df['AQI_Bucket'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
cols = ['NO','NOx','NH3','SO2','O3','Benzene','Toluene']

for col in cols:
    df[col].fillna(df[col].median(), inplace=True)
df.drop(columns=['NH3'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv("aqi_dataset_cleaned.csv", index=False)
```
